Обнаруженные проблемы в данных связаны с обнаруженными несколькими NaN-ами в столбцах продолжительности звонков, испоспользованных мегабайт интернета. Это связанно скорее всего с тем, что клиент в этом месяце не звонил/не использовал интернет/не писал сообщения, а нулевая сумма была записана как NaN. Исправим это.


```python
df_month_general_profit_final['sent_messages_monthly'].fillna(0, inplace=True)
```

```python
def compute_profit(line):
    x = (line['duration_sum'] - line['minutes_included']) * line['rub_per_minute'] # Вычисляем стоимость всех потраченных минут
    y = (line ['gb_used'] - line['mb_per_month_included'] / 1024)  * line['rub_per_gb'] # Затем цену потраченного интернет-трафика
    z = (line['sent_messages_monthly'] - line['messages_included']) * line['rub_per_message'] # И стоимость всех отправленных сообщений
    
    result = x * (x > 0).astype(float) + y * (y > 0).astype(float) + z * (z > 0).astype(float) + line['rub_monthly_fee'] # Домножаем каждую переменную на индикатор, который дает 1 если переменная больше нуля и 0 - если меньше нуля. Таким образом зануляем данные трат в пределах тарифного пакета. 
    return result
```

```python
df_month_general_profit_final['month_profit'] = compute_profit(df_month_general_profit_final)
df_month_final_profit_cuted = df_month_general_profit_final[['user_id', 'month', 'month_profit']]
```

### Проверка гипотез

#### Проверка первой гипотезы

Теперь проверим гипотезы:
1) Cредняя выручка пользователей тарифов «Ультра» и «Смарт» различаются.

Сформулируем нулевую гипотезу H₀:: cредняя прибыль от тарифа ultra и средняя прибыль от тарифа smart - одинакова.  Это двусторонняя гипотеза.

Альтернативная гипотеза: cредняя прибыль от тарифа ultra отличается от средней прибыли от тарифа smart.

Провера гипотезы: 

```python
from scipy import stats as st

alpha = .05 # критический уровень статистической значимости
# если p-value окажется больше него - отвергнем гипотезу

print('Грубая cредняя прибыль таирифа "Ultra": ',month_profit_ultra['month_profit'].mean())
print('Грубая cредняя прибыль таирифа "Smart": ',month_profit_smart['month_profit'].mean())

results = st.ttest_ind(month_profit_smart['month_profit'], month_profit_ultra['month_profit'], )

print('p-value:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

#### Вывод:

Мы отвергаем нулевую гипотезу и принимаем альтернативную: cредняя прибыль от тарифа ultra отличается от средней прибыли от тарифа smart.

#### Проверка второй гипотезы

Проверим следующую гипотезу: 

2) *Средняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов.*

Сформулируем нулевую гипотезу H₀: cредняя прибыль от клиентов из Москвы равна средней прибыли от клиентов из регионов. Это двусторонняя гипотеза.

Альтернативная гипотеза: cредняя прибыль от клиентов из Москвы не равна средней прибыли от клиентов из остальных городов.

```python
display(revenue.head())
```

```python
alpha = .01 # критический уровень статистической значимости выберем большой, т.к. есть все основания полагать, что разница в прибыли в Москве и в других городах имеется
# если p-value окажется больше него - отвергнем гипотезу

results = st.ttest_ind(avg_profit_Moscow, avg_profit_other_cities)

print('p-value:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```
