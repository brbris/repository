### Изменение типов данных

Все явные пропуски заполнены, кроме колонок, которые было решено не заполнять, а именно `living_area`, `locality_name` и `floors_total`.

Нужно будет изменить тип данных следующих столбцов:
* `first_day_exposition` методом date_time()
* `floors_total`, `balcony` - поменять столбец на целочисленный тип, т.к. нули после точки здесь лишние, никто не продает квартиры между этажами или с частью балкона.
* `airports_nearest`, `city_centers_nearest`, `parks_nearest`, `ponds_nearest`, `days_exposition` тоже можно округлить, т.к. расстояния представлены в метрах и вряд ли сантиметры после точки сильно повлияют на наш анализ, как и информация о длительности дней размещения объявления вплоть до часа.
* Данные о количестве парков и прудов в округе содержащие числа после точки  из колонок `ponds_around_3km`, `parks_around_3km` также не имеют смысла, округлим их.


```python

for column in data.columns:
    columns_to_int_type = ['floors_total', 'balcony', 'airports_nearest', 'city_centers_nearest', 'parks_nearest', 'ponds_nearest', 'days_exposition', 'ponds_around_3km', 'parks_around_3km']
    try:
        if column in columns_to_int_type:
            data[column] = data[column].astype(int)
    except:
        print('Что-то не так с данными в колонке', column)
```

Обнаружили проблемную колонку `floors_total`, не позволяющую перевести тип данных в целочисленный. Эта ошибка возникла из-за того, что мы оставили в данных незаполненные NaN-ы. Придется пропуски заполнить :) Пропуск можно заполнить медианным значением из выборки, например.

```python
data['floors_total'] = data['floors_total'].fillna(data['floors_total'].median())
```

Еще раз запустим цикл, чтобы изменить тип данных:

```python
for column in data.columns:
    columns_to_int_type = ['floors_total', 'balcony', 'airports_nearest', 'city_centers_nearest', 'parks_nearest', 'ponds_nearest', 'ponds_around_3km', 'parks_around_3km']
    if column in columns_to_int_type:
        data[column] = data[column].astype(int)
```

Ваша задача — выяснить, какая область входит в центр. Создайте столбец с расстоянием до центра в километрах: округлите до целых значений. После этого посчитайте среднюю цену для каждого километра. Постройте график: он должен показывать, как цена зависит от удалённости от центра. Определите границу, где график сильно меняется — это и будет центральная зона.

```python
piter_data = selected_data[selected_data['locality_name'] == 'Санкт-Петербург'] # Выберем данные по квартирам в Питере
piter_data['to_center'] = (piter_data['city_centers_nearest']/1000).round().astype(int) # Создаю столбец с расстоянием до центра в киллометрах (делю на 1000), округляю до целых и сохраняю в формате int
km_price = pd.pivot_table(piter_data, values=['last_price', 'total_area'], index='to_center') # Создаю в переменной сводную таблицу, которая будет хранить сгруппированные киллометры и среднюю цену по ним
km_price['mean_price'] = km_price['last_price'] / km_price['total_area'] # Найдем среднюю цену за метр для каждого киллометра
km_price['to_center'] = km_price.index # В индексах лежат киллометры до центра, и, поскольку мы не можем построить график от индекса, мы заводим для него столбец
fig = km_price.plot(y='mean_price', x='to_center', style='-', grid=True) # Строим график 
fig.set_xlabel('killometers to center')
fig.set_ylabel('Avg price per meter')
```
